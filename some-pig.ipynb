```python
class Player:
    
    def __init__(self):
        self.rolls_history = []
        
        
    def roll_dice(self):
        return random.choice([1 ,2 ,3 ,4 ,5 ,6])
    

    def decision(self):
        return random.choice(["Hold", "Roll"])
    
    
    def turn(self): 
        roll = self.roll_dice()
        if roll == 1:
            return []
        else:
            return [roll]
        
```

```python
class Simulation:
    
    def __init__(self, player):
        self.player = player
        self.score = 0
        
    def update_score(self, rolls_history):
        total = 0
        for ind in range(len(rolls_history)):
            for ind2 in range(len(rolls_history[ind])):
                total += rolls_history[ind][ind2]
        return total
    
    
    def iterate(self):
        sim_history = []      
        for _ in range(7):
            sim_history.append(self.player.turn())
        self.player.rolls_history = sim_history
        self.score = self.update_score(sim_history)
        
   
    def report(self):
        return self.player.rolls_history, self.score
        
            
```
